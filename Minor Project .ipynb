### Creating piviot table to our requirement for analysis

## Insight from Delhi June 2020 Covid-19

### Percentage distribution Of Covid-19 Infected, Recovered, Deceased as on June 2020 of Delhi

```python
infected = sum(table["Hospitalized"])
deceased = sum(table["Deceased"])
recovered = sum(table["Recovered"])

labels = ['Infected', 'Recovered', 'Deceased']
sizes = [infected, recovered, deceased]
colors = ['lightcoral', 'yellowgreen', 'red']
explode = (0, 0, 0) 
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True)
plt.title("Percentage distribution Of Covid-19 Infected, Recovered, Deceased as on June 2020 of Delhi")
plt.axis('equal')
plt.show()
```

### Relation between Infected people & Deceased personal of Delhi 2020 

```python
index1 = ["06 Jun","13 Jun","14 Jun","15 Jun","16 Jun","17 Jun","18 Jun",
         "19 Jun","20 Jun","21 Jun","22 Jun","23 Jun","24 Jun","25 Jun",
         "26 Jun","27 Jun","28 June","29 Jun","30 Jun"]

x = np.arange(len(index1))

plt.scatter(index1,table["Hospitalized"],label="Infected")
plt.scatter(index1,table["Deceased"],label="Deceased")
plt.title("Relation between Infected people & Deceased personal of Delhi 2020 ")
plt.xticks(x,index1,rotation=55)
plt.ylabel("Number of Deceased people")
plt.legend()
plt.show()
```

### Relation between Infected , Recovered & Deceased personal of Delhi 2020

```python
x = np.arange(len(index1))

plt.plot(index1,table["Hospitalized"],label="Infected")
plt.plot(index1,table["Recovered"],label="Recovered")
plt.plot(index1,table["Deceased"],label="Deceased")
plt.title("Relation between Infected , Recovered & Deceased personal of Delhi 2020 ")
plt.ylabel("Number of  people")
plt.xticks(x,index1,rotation=55)
plt.legend()
plt.show()
```

## COVID-19 Recovery rate Calculation Weekly for Delhi June 2020  

```python
week1 = table.iloc[0]
week2 = table.iloc[1] 
week3 = table.iloc[2:9]
week4 = table.iloc[9:16]
week5 = table.iloc[16:19]

week1R=week1['Recovered']
week1I=week1['Hospitalized']

week2R=week2['Recovered']
week2I=week2['Hospitalized']

week3R=week3['Recovered'].iloc[0]
week3I=week3['Hospitalized'].iloc[0]

week4R=week4['Recovered'].iloc[0]
week4I=week4['Hospitalized'].iloc[0]

week5R=week5['Recovered'].iloc[0]
week5I=week5['Hospitalized'].iloc[0]

RR_weeks = []


recovered = week1R
infected = week1I
recovery_rate = (recovered/infected) * 100
RR_weeks.append(recovery_rate)
        
        
recovered=week2R
infected = week2I
recovery_rate = (recovered/infected) * 100
RR_weeks.append(recovery_rate)
        
        
recovered=week3R
infected = week3I
recovery_rate = (recovered/infected) * 100
RR_weeks.append(recovery_rate) 
  
        

recovered=week4R
infected = week4I
recovery_rate = (recovered/infected) * 100
RR_weeks.append(recovery_rate)
        
        

recovered=week5R
infected = week5I
recovery_rate = (recovered/infected) * 100
RR_weeks.append(recovery_rate)
       
        
        
RR_weeks
```

## COVID-19 Mortality rate Calculation weekly for Delhi June 2020

```python

week1 = table.iloc[0]
week2 = table.iloc[1] 
week3 = table.iloc[2:9]
week4 = table.iloc[9:16]
week5 = table.iloc[16:19]

week1D=week1['Deceased']
week1I=week1['Hospitalized']

week2D=week2['Deceased']
week2I=week2['Hospitalized']

week3D=week3['Deceased'].iloc[0]
week3I=week3['Hospitalized'].iloc[0]

week4D=week4['Deceased'].iloc[0]
week4I=week4['Hospitalized'].iloc[0]

week5D=week5['Deceased'].iloc[0]
week5I=week5['Hospitalized'].iloc[0]

MR_weeks = []


deceased = week1D
infected = week1I
mortality_rate = (deceased/infected) * 100
MR_weeks.append(mortality_rate)
        
        
deceased=week2D
infected = week2I
mortality_rate = (deceased/infected) * 100
MR_weeks.append(mortality_rate)
        
        
deceased=week3D
infected = week3I
mortality_rate = (deceased/infected) * 100
MR_weeks.append(mortality_rate) 
  
        

deceased=week4D
infected = week4I
mortality_rate = (deceased/infected) * 100
MR_weeks.append(mortality_rate)
        
        

deceased=week5D
infected = week5I
mortality_rate = (deceased/infected) * 100
MR_weeks.append(mortality_rate)
       
        
        
MR_weeks
```

### Relation between Recovery Rate & Mortality Rate of Delhi June weekly 2020

```python
index2 = ["1-6 June","7-13 June","14-20 June","21-27 June","28-39 June"]
x = np.arange(len(index2))

plt.plot(index2,RR_weeks,label="Recovery Rate")
plt.plot(index2,MR_weeks,label="Mortality Rate ")

plt.title("Relation between Recovery Rate & Mortality Rate of Delhi June weekly 2020 ")
plt.ylabel("Percentage %")
plt.xticks(x,index2,rotation=55)
plt.legend()
plt.show()
```
